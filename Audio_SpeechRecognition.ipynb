```python
from transformers import pipeline

# Initialize ASR pipeline with a different model
asr_pipeline = pipeline('automatic-speech-recognition', model='facebook/wav2vec2-base-960h', device=-1)

def transcribe_audio(audio_file_path):
    # Transcribe audio
    transcriptions = asr_pipeline(audio_file_path)
    return transcriptions

def display_audio(audio_file_path):
    # Display audio for listening
    from IPython.display import Audio, display
    display(Audio(filename=audio_file_path, autoplay=True))

def main():
    audio_file_path = '/content/you_got_a_text_message.mp3'

    # Display audio for listening
    display_audio(audio_file_path)

    # Transcribe audio
    transcriptions = transcribe_audio(audio_file_path)
    print("Transcription:", transcriptions)

if __name__ == "__main__":
    main()
```
